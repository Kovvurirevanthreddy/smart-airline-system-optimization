```python
def dijkstra(graph,start,end):
    dist={node:float('inf') for node in graph}
    prev={}
    dist[start]=0
    pq=[(0,start)]

    while pq:
        d,node=heapq.heappop(pq)

        if node not in graph:
            continue

        if node==end:
            break

        for n,w in graph[node].items():
            nd=d+w
            if nd<dist[n]:
                dist[n]=nd
                prev[n]=node
                heapq.heappush(pq,(nd,n))

    path=[]
    curr=end

    if end not in prev and end!=start:
        return [],float('inf')

    while curr!=start:
        path.append(curr)
        curr=prev[curr]
    path.append(start)
    path.reverse()

    return path,dist[end]
```
